```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import make_regression
from sklearn.datasets import make_classification

from linearRegression import LinearRegression
from logisticRegression import LogisticRegression
```

```python
X, y, coeffs = make_regression(n_samples=50, n_features=1, n_informative=1,
                               noise=20, coef=True, random_state=1)
model = LinearRegression(verbose=False)
model.fit(X, y)

predictions = model.predict(X)

plt.scatter(X, y, color='blue', label='Data')
plt.plot(X, predictions, color='red', label='Predictions')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Linear Regression')
plt.legend()
plt.show()
```

```python
X, y = make_classification(n_samples=50, n_features=1, n_informative=1,
                           n_redundant=0, random_state=1, n_clusters_per_class=1,
                           class_sep=0.4)

#   
model = LogisticRegression(verbose=True)
model.fit(X, y)

#   
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

#     (z = X.dot(weights))
z = np.column_stack([np.ones((X.shape[0])), X]).dot(model.weights)
probabilities = sigmoid(z)

#  
plt.scatter(X[y == 0], y[y == 0], color='blue', label='Class 0')
plt.scatter(X[y == 1], y[y == 1], color='red', label='Class 1')

#   
X_sorted = np.sort(X, axis=0)
z_sorted = np.column_stack([np.ones((X_sorted.shape[0])), X_sorted]).dot(model.weights)
prob_sorted = sigmoid(z_sorted)
plt.plot(X_sorted, prob_sorted, color='green', label='Sigmoid curve')

plt.xlabel('X')
plt.ylabel('Predicted probability')
plt.title('Logistic Regression Sigmoid Curve')
plt.legend()
plt.show()
```
